```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
df.shape
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
# Drop blank columns
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df.describe()
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

Exploratory Data Analysis

Gender

```python
df.columns
```

```python
ax = sns.countplot(data=df, x= 'Gender', palette='Set2')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.barplot(x = 'Gender', y='Amount', data = sales_gen, palette='Set2')
plt.show()
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men.*

Age

```python
ax = sns.countplot(data=df, x= 'Age Group', hue = 'Gender', palette='Set2')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.barplot(x = 'Age Group', y='Amount', data = sales_age, palette='Set2')
plt.show()
```

*From above graphs we can see that most of the buyers are of age group 26-35 yr female*

State

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(17,5)})
sns.barplot(data = sales_state, x = 'State', y='Orders', palette='Set2')
plt.show()
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(17,5)})
sns.barplot(data = sales_state, x = 'State', y='Amount', palette='Set2')
plt.show()
```

*From above graphs we can see that most of the orders are from Uttar Pradesh, Maharashtra and Karnataka respectively but total sales/ amount is for Haryana is higher than Kerela.*

Marital Status

```python
ax = sns.countplot(data=df, x= 'Marital_Status', palette='Set2')

sns.set(rc = {'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_married = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(7,5)})
sns.barplot(data = sales_married, x = 'Marital_Status', y='Amount', hue = 'Gender', palette='Set2')
plt.show()
```

*From above graphs we can see that most of the buyers are married(women) and they have high purchasing power*

Occupation

```python
sns.set(rc = {'figure.figsize':(20,5)})
ax = sns.countplot(data=df, x= 'Occupation', palette='Set2')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sns.set(rc = {'figure.figsize':(20,5)})
sales_work = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.barplot(data = sales_work, x = 'Occupation', y='Amount', palette='Set2')
plt.show()
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare ans Aviation sector. And least purchasing buyers are from Agriculture sector.*

Product Category

```python
sns.set(rc = {'figure.figsize':(25,5)})
ax = sns.countplot(data=df, x= 'Product_Category', palette='Set2')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sns.set(rc = {'figure.figsize':(25,5)})
sales_prod_cat = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)

sns.barplot(data = sales_prod_cat, x = 'Product_Category', y='Amount', palette='Set2')
plt.show()
```

*From above graphs we can see that most of the sold product are from Food, Clothing and Electronics category*

```python
sns.set(rc = {'figure.figsize':(20,5)})
sales_prod = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)

sns.barplot(data = sales_prod, x = 'Product_ID', y='Orders', palette='Set2')
plt.show()
```

Conclusion :

*Married women age group 26-35 yrs from UP, Maharashtra and Karnataka working in IT, Healthcare and Aviation are most likely to buy products from Food, Clothing and Electronics Category*
